```python
# IMPORTS
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from ID3_paper import ID3DecisionTree # class from paper
from sklearn.tree import DecisionTreeClassifier
```

```python
# Read in the data
data = pd.read_csv("Dataset/breast-cancer.csv")
data.head()
```

```python
# Verify binarized labels
print(data["diagnosis"].value_counts()) # -- good --
data.info() # no missing values -- good --
```

```python
# Get our training and testing datasets
X = data.drop('diagnosis', axis=1).values
y = data['diagnosis'].apply(lambda x: 1 if x == 'M' else 0).values

X_train, y_train = X[:450], y[:450]
X_test, y_test = X[450:], y[450:]
```

```python
# Test the ID3 Decision Tree class
id3_tree = ID3DecisionTree()
id3_tree.fit(X_train, y_train)
id3_preds = id3_tree.predict(X_test)
id3_accuracy = accuracy_score(y_test, id3_preds)
print(f"ID3 Accuracy Score: {id3_accuracy}")
```

```python
# Test Sklearn
d_tree = DecisionTreeClassifier()
d_tree.fit(X_train, y_train)
sk_preds = d_tree.predict(X_test)
sk_accuracy = accuracy_score(y_test, sk_preds)
print(f"Sklearn Accuracy Score: {sk_accuracy}")
```
